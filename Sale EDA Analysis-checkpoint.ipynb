```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SalesDataset.csv')
```

```python
df.head(), df.info()
```

```python
# Drop the 'Unnamed: 0' column as it's just an index
df_cleaned = df.drop(columns=['Unnamed: 0'])

# Convert 'Date' column to datetime format
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], errors='coerce')

# Check for missing values
missing_values = df_cleaned.isnull().sum()

# Display cleaned data types and missing values
df_cleaned.dtypes, missing_values
```

```python
# Step 2: Descriptive Statistics

# Summary statistics for numerical columns
numeric_summary = df_cleaned.describe()

# Count of unique values for categorical columns
categorical_summary = df_cleaned.select_dtypes(include='object').nunique()

# Display both summaries
numeric_summary, categorical_summary
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# 1. Gender-wise Total Sales
plt.figure()
sns.barplot(data=df_cleaned, x='Gender', y='Total Amount', estimator=sum)
plt.title('Total Sales by Gender')
plt.ylabel('Total Sales')
plt.xlabel('Gender')
plt.tight_layout()
plt.show()
```

```python
# 2. Product Category-wise Sales
plt.figure()
sns.barplot(data=df_cleaned, x='Product Category', y='Total Amount', estimator=sum)
plt.title('Total Sales by Product Category')
plt.ylabel('Total Sales')
plt.xlabel('Product Category')
plt.tight_layout()
plt.show()
```

```python
# 3. Age Distribution
plt.figure()
sns.histplot(df_cleaned['Age'], bins=15, kde=True)
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# 4. Monthly Sales Trend
df_cleaned['Month'] = df_cleaned['Date'].dt.to_period('M').astype(str)
monthly_sales = df_cleaned.groupby('Month')['Total Amount'].sum().reset_index()

plt.figure()
sns.lineplot(data=monthly_sales, x='Month', y='Total Amount', marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
# 5. Correlation Heatmap
plt.figure()
sns.heatmap(df_cleaned[['Age', 'Quantity', 'Price per Unit', 'Total Amount']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Define age groups for segmentation
bins = [18, 25, 35, 45, 55, 65]
labels = ['18-25', '26-35', '36-45', '46-55', '56-65']
df_cleaned['Age Group'] = pd.cut(df_cleaned['Age'], bins=bins, labels=labels, right=True)

# Group by Age Group and Product Category
segment_product = df_cleaned.groupby(['Age Group', 'Product Category'])['Total Amount'].sum().reset_index()

# Visualization: Heatmap of sales across Age Group and Product Category
pivot_table = segment_product.pivot(index='Age Group', columns='Product Category', values='Total Amount')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Customer Segmentation vs Product Category (Total Sales)')
plt.xlabel('Product Category')
plt.ylabel('Age Group')
plt.tight_layout()
plt.show()
```
